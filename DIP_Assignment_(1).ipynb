```python
def normalization(img):
  min_val = 255
  for i in img:
    for v in i:
        if min_val > v:
            min_val = v
  fm = []
  max_fm = -1
  for i in img:
    img = []    
    for pixel in i:
        val = pixel - min_val
        img.append(val)
        
        if (val > max_fm):
            max_fm = val
    fm.append(img)

  k = 256-1 #k=255

  fr=[]
  for i in fm:
    img = []
    for pixel in i:
      val = k * (pixel/max_fm)
      img.append(val)
    fr.append(img)

  return fr
```

```python
 def bitPlane(image,bit_number = 0):
        if bit_number >= 8:
            print("Bit Plane number is wrong")
            return None 
        bit_image_values = []
        for i in range(len(image)):
            bit_image_values.append(image[i])
        final_image_values = image
        
        for j in range(bit_number+1):
            bit_list = [  int(i[j] )    for i in bit_image_values]
            temp_values = np.array( bit_list ) * (2^(7-j))
            for i in range(len(temp_values)):
                final_image_values[i] += temp_values[i]

        return final_image_values
```

```python
def brightness(image):
        image_values_sum = 0
        for j in image:
          for i in j:
            image_values_sum += i
        dim = dimension(image)
        dim = list(dim)
        length = dim[0]
        width = dim[1]
        return image_values_sum / (length *width)
```

```python
def contrastValue(image):
        bright = brightness(image)
        image_values_sum_squares = 0
        for j in image:
          for i in j:
            image_values_sum_squares += (bright  - i) ** 2

        dim = dimension(image)
        dim = list(dim)
        length = dim[0]
        width = dim[1]
        final_result = ((image_values_sum_squares/ (length*width ** 0.5)))
        return final_result
```
